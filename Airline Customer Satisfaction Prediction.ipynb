# üëâüèª Apllying Multiple Models Together for the best Accuracy:-

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
lis=[LinearRegression,SVC,DecisionTreeClassifier,RandomForestClassifier,KNeighborsClassifier]
```

```python
k=KFold(n_splits=4)
cross_val_score(LinearRegression(),x_train,y_train,cv=k)
```

```python
for i in lis:
    model=i()
    k=KFold(n_splits=4)
    print(i)
    print(cross_val_score(model,x_train,y_train,cv=k).mean())
```

```python
model=RandomForestClassifier(n_estimators=250,random_state=20,max_depth=9)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
dic={"n_estimators":[150,200,250],
    "random_state":[10,20,30],
    "max_depth":[7,8,9]}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random=RandomizedSearchCV(RandomForestClassifier(),param_distributions=dic,cv=k,verbose=2,n_iter=10)
```

```python
random.fit(x_train,y_train)
```

```python
random.best_params_
```

```python
model=random.best_estimator_
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
